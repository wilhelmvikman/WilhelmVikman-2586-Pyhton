# Task 1: A list of dictionaries

```python

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# Quick check
print(transactions)
```

### a) Filter invalid transactions

```python


valid_transactions = []

for transaction in transactions:
    # Keep only valid ones
    if transaction["price"] > 0 and transaction["quantity"] > 0:
        valid_transactions.append(transaction)

# Check result
print(valid_transactions)
```

### b) Compute Total per transaction

```python
# Add total = price * quantity to each valid transaction

for transaction in valid_transactions:
  transaction["total"] = transaction["price"] * transaction["quantity"]

# Check result
print(valid_transactions)
```

### c) Group valid transactions by region and compute total revenue

```python
# Create a dictonary to store total revenue per region
revenue_by_region = {}
```

```python
# Loop through transactions and sum totals by region

for transaction in valid_transactions:
  region = transaction["region"]
  total = transaction["total"]

  # If region is not yet in dictionary, initialize it
  if region not in revenue_by_region:
    revenue_by_region[region] = 0

  # Add total to region
  revenue_by_region[region] += total

# Check result
print(revenue_by_region)
```

### d) Print the region with the highest total revenue and its value using an f-string.

```python
# Find the region with the highest revenue

max_region = None
max_revenue = 0

for region, revenue in revenue_by_region.items():
  if revenue > max_revenue:
    max_revenue = revenue
    max_region = region

# Check result
print("Region with highest revenue:", max_region)
print("Revenue:", max_revenue)
```

# Task 2: Creating a function

```python
# List of daily sales
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

print(sales)
```

### creating a function that summarizes sales

```python
def summarize_sales(sales_list):
  # Total sales
  total = sum(sales_list)

  # Average sales (rounded to two decimals)
  average = round(total / len(sales_list), 2)

  # Best day (starting from 1)
  best_day = sales_list.index(max(sales_list)) + 1

  # Worst day (starting from 1)
  worst_day = sales_list.index(min(sales_list)) + 1

  # Days above average sales
  above_average_count = 0
  for value in sales_list:
    if value > average:
      above_average_count += 1


  return {
      "total": total,
      "average": average,
      "best_day": best_day,
      "worst_day": worst_day,
      "above_average_count": above_average_count
  }
```

```python
result = summarize_sales(sales)
print(result)
```

# Task 3: Loops

```python
# List of student records
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

print(students)
```

### a) Create a new list called passing that contains only the names of students with a grade of 50 or above. Print the list.

```python
# Create a list of a passing students

passing = []

for student in students:
  if student["grade"] >= 50:
    passing.append(student["name"])

print(passing)
```

### b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print "[name]: Excellent"; if between 50 and 79, print "[name]: Pass"; otherwise print "[name]: Fail".

```python
# Loop through the students and print status message

for student in students:
  name = student["name"]
  grade = student["grade"]

  if grade >= 80:
    print(f"{name}: Excellent")
  elif grade >= 50:
    print(f"{name}: Pass")
  else:
    print(f"{name}: Fail")
```

### c) Count how many students have an attendance rate below 0.80. Print the count.

```python
low_attendance_count = 0

for student in students:
  if student["attendance"] < 0.80:
    low_attendance_count += 1

print(low_attendance_count)

```

d) Find the student with the highest grade using a loop (do not use max()). Print the student's name and grade.

```python
top_student = None
highest_grade = 0

for student in students:
  if student["grade"] > highest_grade:
    highest_grade = student["grade"]
    top_student = student["name"]

print(f"Top_student: {top_student}, Grade: {highest_grade}")

```

e) Find the student with the highest grade using max(). Print the student's name and grade.

```python
top_student = max(students, key=lambda student: student["grade"])

print(f"Top student: {top_student['name']}, Grade: {top_student['grade']}")
```

# Task 4: Class and methods

```python
# Design a class called Portfolio that represents an investment portfolio. The constructor is provided for you:
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

# a)

    def add_asset(self, ticker, shares, price_per_share):
      # Add or update asset
      self.holdings[ticker] = {
          "shares": shares,
          "price_per_share": price_per_share
      }

# b)

    def total_value(self):
      total = 0

      for asset in self.holdings.values():
        shares = asset["shares"]
        price = asset["price_per_share"]
        total += shares * price

      return total

# c)

    def most_valuable(self):
      max_value = 0
      top_ticker = None

      for ticker, asset in self.holdings.items():
        value = asset["shares"] * asset["price_per_share"]

        if value > max_value:
          max_value = value
          top_ticker = ticker

      return top_ticker

# d)

    def __str__(self):
      num_assets = len(self.holdings)
      total = self.total_value()

      return f"Portfolio owner: {self.owner}, Assets: {num_assets}, Total value: {total}"
```

a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

```python
# Portfolio owned by Wilhelm containing the following assets:

portfolio = Portfolio("Wilhelm")

# Assets
portfolio.add_asset("AAPL", 10, 150)
portfolio.add_asset("GOOGL", 5, 350)
portfolio.add_asset("NVDA", 15, 160)

```

# Task 5: Importing a module

```python
# Order dates

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

```python
print(orders)
```

```python
from datetime import datetime
```

```python
# Converting date strings into datetime.date objects

dates = []

for order in orders:
  date_str = order[0]

  date_obj = datetime.strptime(date_str, "%Y-%m-%d").date()

  dates.append(date_obj)

  print(dates)
```

```python
# Find the earliest and latest dates

earliest = min(dates)
latest = max(dates)

# Compute the difference
difference = (latest - earliest).days

# Result: Number of days between earliest and latest
print(f"Earliest date: {earliest}")
print(f"Latest date: {latest}")
print(f"Days between: {difference}")
```

# Task 8: More Pandas

```python
# Script of daily website traffic

import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

### a) Create a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value. Set date as the index.

```python
# Create new DataFrame
df_traffic = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

# setting "date" as index
df_traffic = df_traffic.set_index("date")

print(df_traffic.head())
```

### b) Compute two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value).



```python
# Compute conversions (rounded to the nearest integer)
df_traffic["conversions"] = (df_traffic["visitors"] * df_traffic["conversion_rate"]).round(0)

# Compute daily_revenue
df_traffic["daily_revenue"] = df_traffic["conversions"] * df_traffic["avg_order_value"]

print(df_traffic.head())
```
